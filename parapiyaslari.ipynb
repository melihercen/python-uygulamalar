```python
import json
import datetime
import http.client

#Apileri collectapi.com sitesinden alındı
#Döviz Kurları ve Hisse senetlerinin kodlarını giriniz örnek "USD","TRY","THYAO","AKBNK" .Kriptoda tam ad giriniz 
class Parapiyasaları():
    def istanbul_borsası():
        # HTTP bağlantısını oluştur
        conn = http.client.HTTPSConnection("api.collectapi.com")

        # HTTP başlıklarını ayarla
        headers = {
            'content-type': "application/json",
            'authorization': "buraya api keyinizi girin"
        }

        # Kullanıcıdan hedef hisse senedini al
        hisse_sembol = input("Hedef Hisse senedi: ").upper()

        # API'ye GET isteğini yap
        conn.request("GET", f"/economy/hisseSenedi", headers=headers)

        # API'den yanıtı al
        res = conn.getresponse()
        data = res.read()

        # JSON yanıtını Python sözlüğüne dönüştür
        hisse_dict = json.loads(data)

        # 'result' anahtarına sahip olan kısmı al
        results = hisse_dict.get('result', [])

        #Güncel tarihi öğrenme
        date=datetime.datetime.now()
        date_day=date.strftime("%A")
        print(date_day)
        
        if date_day=="Saturday" or date_day=="Sunday":
                print("Piyasalar bugün kapalı\nCuma günundeki veriler goruyorsunuz")
        #for dönüsü ile listede gezinerek istediğimiz sonuca ulaşma
        for result in results:
            if result.get("code")==hisse_sembol:
                hisse_fiyati = result.get("lastprice")
                hisse_min=result.get("min")
                hisse_max=result.get("max")
                print(f"{hisse_sembol} hissesinin fiyatı: {hisse_fiyati}\nGün içinde\nMin:{hisse_min}\nMax:{hisse_max}")
                print("-----------------------------------")
                
                break

        else:
            print("Hisse bulunamadı.")
        
    def Dovizcevir():
        conn = http.client.HTTPSConnection("api.collectapi.com")

        headers = {
            'content-type': "application/json",
            'authorization': "aburaya api keyinizi girin"
            }

        code=input("İstenilen doviz türü :").upper()
        base=input("Eldeki doviz türü: ").upper()
        tutar=input("Çevrilecek tutar:")
        conn.request("GET", f"/economy/currencyToAll?int={tutar}&base={base}", headers=headers)

        res = conn.getresponse()
        data = res.read()

        doviz_dict=json.loads(data)

        results=doviz_dict.get("result",{}).get("data",[])


        for result in results:
            if result.get("code")==code:
                sonuc=result.get("calculated")
                print(f"{sonuc} {code}")
                print("-----------------------------------")
                
                break
        else:
            print("hata")
            
    def KriptoBorsasi():
        conn = http.client.HTTPSConnection("api.collectapi.com")

        headers = {
            'content-type': "application/json",
            'authorization': "buraya api keyinizi girin"
            }

        conn.request("GET", "/economy/cripto", headers=headers)
        kripto_ad=input("Kripto hisse adını girin: ").capitalize()
        res = conn.getresponse()
        data = res.read()
        kripto_dict=json.loads(data)

        results=kripto_dict.get("result",[])

        for result in results:
            if result.get("name")==kripto_ad:
                sembol=result.get("code")
                deger=result.get("price")
                gun=result.get("changeDay")
                hafta=result.get("changeWeek")
                saat=result.get("changeHour")
                print(f"{sembol}\n{deger} Usd\nSaatlik değişim: {saat}\nGünlük Değişim: {gun}\nHaftalık Değişim: {hafta}")
                print("-----------------------------------")
                
                break
        else:
            print("Hata")
            

            
            
    def doviz_kur():
        conn = http.client.HTTPSConnection("api.collectapi.com")
        
        headers = {
            'content-type': "application/json",
            'authorization': "buraya api keyinizi girin"
            }

        conn.request("GET", "/economy/allCurrency", headers=headers)
        istenen_kur=input("Görmek istediğiniz kur: ").upper()
        
        res = conn.getresponse()
        data = res.read()
        kur_dict=json.loads(data)

        results=kur_dict.get("result",[])

        for result in results:
            if result.get("code")==istenen_kur:
                sembol=result.get("code")
                alis=result.get("buying")
                satis=result.get("selling")
                print(f"{alis}TL=1{sembol}\nAlış={alis}\nSatış={satis}")
                print("-----------------------------------")
                break
        else:
            print("Hata")
            

while True:
    x = datetime.datetime.now()
    print(x.strftime("%c"))
    print("1-Isatnbul Borsası Piyasaları")
    print("2-Doviz Cevirici")
    print("3-Kripto Borsası Piyasaları")
    print("4-Doviz Kurları TL bazında")
    print("5-Çıkış")
    secim=input("Yapmak istediginiz islemi secınız 1/2/3/4/5")
    if secim=="1":
        Parapiyasaları.istanbul_borsası()
    elif secim == "2":
        Parapiyasaları.Dovizcevir()
    elif secim == "3":
        Parapiyasaları.KriptoBorsasi()
    elif secim == "4":
        Parapiyasaları.doviz_kur()
    elif secim=="5":
        break
    else:
        print("Hatalı seçim. Lütfen tekrar deneyin.")
    
            
            

        
        


```
